```python
from datetime import datetime, timedelta
import datetime as dt


def current_price(df):
    max_dates = df.groupby('ticker')['date_extract'].max()
    max_values = pd.merge(df, max_dates, on=['ticker','date_extract'], how='inner')
    df = max_values[['ticker', 'close','volume','date_extract']]
    df = df.rename(columns={'close': 'cp_close', 'volume': 'cp_volume','date_extract': 'cp_date_extract'})
    return df

def one_month_price(df):
    df['date_extract'] = pd.to_datetime(df['date_extract'])
    grouped = df.groupby('ticker',as_index=False)
    sorted_groups = grouped.apply(lambda x: x.sort_values('date_extract', ascending=False))
    sorted_groups['date_extract'] = pd.to_datetime(sorted_groups['date_extract']).dt.date
    df= sorted_groups.groupby('ticker',as_index=False).nth(26)
    df = df[['ticker', 'close','volume','date_extract']]
    df = df.rename(columns={'close': 'one_m_close', 'volume': 'one_m_volume','date_extract': 'one_m_date_extract'})
    return df

def three_month_price(df):
    df['date_extract'] = pd.to_datetime(df['date_extract'])
    grouped = df.groupby('ticker',as_index=False)
    sorted_groups = grouped.apply(lambda x: x.sort_values('date_extract', ascending=False))
    sorted_groups['date_extract'] = pd.to_datetime(sorted_groups['date_extract']).dt.date
    df= sorted_groups.groupby('ticker',as_index=False).nth(66)
    df = df[['ticker', 'close','volume','date_extract']]
    df = df.rename(columns={'close': 'three_m_close', 'volume': 'three_m_volume','date_extract': 'three_m_date_extract'})
    return df

def six_month_price(df):
    df['date_extract'] = pd.to_datetime(df['date_extract'])
    grouped = df.groupby('ticker',as_index=False)
    sorted_groups = grouped.apply(lambda x: x.sort_values('date_extract', ascending=False))
    sorted_groups['date_extract'] = pd.to_datetime(sorted_groups['date_extract']).dt.date
    df= sorted_groups.groupby('ticker',as_index=False).nth(132)
    df = df[['ticker', 'close','volume','date_extract']]
    df = df.rename(columns={'close': 'six_m_close', 'volume': 'six_m_volume','date_extract': 'six_m_date_extract'})
    return df

def one_year_price(df):
    df['date_extract'] = pd.to_datetime(df['date_extract'])
    grouped = df.groupby('ticker',as_index=False)
    sorted_groups = grouped.apply(lambda x: x.sort_values('date_extract', ascending=False))
    sorted_groups['date_extract'] = pd.to_datetime(sorted_groups['date_extract']).dt.date
    df= sorted_groups.groupby('ticker',as_index=False).nth(261)
    df = df[['ticker', 'close','volume','date_extract']]
    df = df.rename(columns={'close': 'one_yr_close', 'volume': 'one_yr_volume','date_extract': 'one_yr_date_extract'})
    return df

def three_year_price(df):
    df['date_extract'] = pd.to_datetime(df['date_extract'])
    grouped = df.groupby('ticker',as_index=False)
    sorted_groups = grouped.apply(lambda x: x.sort_values('date_extract', ascending=False))
    sorted_groups['date_extract'] = pd.to_datetime(sorted_groups['date_extract']).dt.date
    df= sorted_groups.groupby('ticker',as_index=False).nth(552)
    df = df[['ticker', 'close','volume','date_extract']]
    df = df.rename(columns={'close': 'three_yr_close', 'volume': 'three_yr_volume','date_extract': 'three_yr_date_extract'})
    return df

def two_year_price(df):
    df['date_extract'] = pd.to_datetime(df['date_extract'])
    grouped = df.groupby('ticker',as_index=False)
    sorted_groups = grouped.apply(lambda x: x.sort_values('date_extract', ascending=False))
    sorted_groups['date_extract'] = pd.to_datetime(sorted_groups['date_extract']).dt.date
    df= sorted_groups.groupby('ticker',as_index=False).nth(261)
    df = df[['ticker', 'close','volume','date_extract']]
    df = df.rename(columns={'close': 'two_yr_close', 'volume': 'two_yr_volume','date_extract': 'two_yr_date_extract'})
    return df

def bottom_price(df):
    df['date_extract'] = pd.to_datetime(df['date_extract'])
    grouped = df.groupby('ticker',as_index=False)
    sorted_groups = grouped.apply(lambda x: x.sort_values('date_extract', ascending=False))
    sorted_groups['date_extract'] = pd.to_datetime(sorted_groups['date_extract']).dt.date
    df= sorted_groups.groupby('ticker',as_index=False).nth(-1)
    df = df[['ticker', 'close','volume','date_extract']]
    df = df.rename(columns={'close': 'bt_close', 'volume': 'bt_volume','date_extract': 'bt_date_extract'})
    return df

def merge_price_fluxuations(df):
    cp = current_price(df)
    one_m = one_month_price(df)
    three_mon = three_month_price(df)
    six_mon = six_month_price(df)
    one_year = one_year_price(df)
    two_year = two_year_price(df)
    three_year = three_year_price(df)
    bottom = bottom_price(df)
    df = pd.merge(cp, one_m, on='ticker', how='outer')
    df = pd.merge(df, three_mon, on='ticker', how='outer')
    df = pd.merge(df, six_mon, on='ticker', how='outer')
    df = pd.merge(df, one_year, on='ticker', how='outer')
    df = pd.merge(df, two_year, on='ticker', how='outer')
    df = pd.merge(df, three_year, on='ticker', how='outer')
    df = pd.merge(df, bottom, on='ticker', how='outer')
    return df
```

```python
def price_analysis_Data(df):
    analysis_Data_frame = pd.DataFrame()
    distinct_values = df['ticker'].unique().tolist()
    for i in distinct_values:
        df_tick = df.loc[df['ticker'] == i].copy()  # Add .copy() to avoid SettingWithCopyWarning
        df_tick.loc[:, 'one_m_price_diff'] = ((df_tick['cp_close'] - df_tick['one_m_close']) / df_tick['cp_close']) * 100
        df_tick.loc[:, 'one_m_negative'] = df_tick['one_m_price_diff'] < 0
        df_tick.loc[:, 'one_mont_vol_diff'] = ((df_tick['cp_volume'] - df_tick['one_m_volume']) / df_tick['cp_volume']) * 100

        df_tick.loc[:, 'three_m_price_diff'] = ((df_tick['cp_close'] - df_tick['three_m_close']) / df_tick['cp_close']) * 100
        df_tick.loc[:, 'three_m_negative'] = df_tick['three_m_price_diff'] < 0
        df_tick.loc[:, 'three_mont_vol_diff'] = ((df_tick['three_m_volume'] - df_tick['cp_volume']) / df_tick['cp_volume']) * 100

        df_tick.loc[:, 'six_m_price_diff'] = ((df_tick['cp_close'] - df_tick['six_m_close']) / df_tick['cp_close']) * 100
        df_tick.loc[:, 'six_m_negative'] = df_tick['six_m_price_diff'] < 0
        df_tick.loc[:, 'six_mont_vol_diff'] = ((df_tick['six_m_volume'] - df_tick['cp_volume']) / df_tick['six_m_volume']) * 100

        df_tick.loc[:, 'one_yr_price_diff'] = ((df_tick['cp_close'] - df_tick['one_yr_close']) / df_tick['cp_close']) * 100
        df_tick.loc[:, 'one_yr_negative'] = df_tick['one_yr_price_diff'] < 0
        df_tick.loc[:, 'one_yr_vol_diff'] = ((df_tick['one_yr_volume'] - df_tick['cp_volume']) / df_tick['one_yr_volume']) * 100

        df_tick.loc[:, 'two_yr_price_diff'] = ((df_tick['cp_close'] - df_tick['two_yr_close']) / df_tick['cp_close']) * 100
        df_tick.loc[:, 'two_yr_negative'] = df_tick['two_yr_price_diff'] < 0
        df_tick.loc[:, 'two_yr_vol_diff'] = ((df_tick['two_yr_volume'] - df_tick['cp_volume']) / df_tick['two_yr_volume']) * 100

        df_tick.loc[:, 'three_yr_price_diff'] = ((df_tick['cp_close'] - df_tick['three_yr_close']) / df_tick['cp_close']) * 100
        df_tick.loc[:, 'three_yr_negative'] = df_tick['three_yr_price_diff'] < 0
        df_tick.loc[:, 'three_yr_vol_diff'] = ((df_tick['three_yr_volume'] - df_tick['cp_volume']) / df_tick['three_yr_volume']) * 100

        df_tick.loc[:, 'bt_price_diff'] = ((df_tick['cp_close'] - df_tick['bt_close']) / df_tick['cp_close']) * 100
        df_tick.loc[:, 'bt_negative'] = df_tick['bt_price_diff'] < 0
        df_tick.loc[:, 'bt_vol_diff'] = ((df_tick['bt_volume'] - df_tick['cp_volume']) / df_tick['bt_volume']) * 100

        analysis_Data_frame = pd.concat([analysis_Data_frame, df_tick], ignore_index=True)

    return analysis_Data_frame
```

```python
# Read the data from the database into a dataframe
income = pd.read_sql_query("SELECT ticker,TotalRevenue,asOfDate,NetIncome,ResearchAndDevelopment,BasicEPS FROM income_quarterly_data where periodType != 'TTM'", conn)
income

def income_quarterly_Values_cagr(df):
    income_quarterly_dataframe = pd.DataFrame(columns=['cagr_revenue_qr', 'ticker','cagr_net_income_qr','cagr_eps_value_qr'])
    distinct_values = df['ticker'].unique().tolist()
    for i in distinct_values:
        df_tick = df.loc[df['ticker'] == i]
        grouped = df_tick.groupby('ticker',group_keys=False)
        sorted_groups = grouped.apply(lambda x: x.sort_values('asOfDate', ascending=True))
        beginning_value = sorted_groups['TotalRevenue'].iloc[0]
        ending_value = sorted_groups['TotalRevenue'].iloc[-1]
        filtered_df = df[df['TotalRevenue'].notnull()]
        num_years = len(filtered_df)
        cagr_rev = (ending_value / beginning_value) ** (1 / num_years) - 1
        
        beginning_value_net = sorted_groups['NetIncome'].iloc[0]
        ending_value_net = sorted_groups['NetIncome'].iloc[-1]
        filtered_df1 = df[df['NetIncome'].notnull()]
        num_years1 = len(filtered_df1)
        cagr_net_in = (ending_value_net / beginning_value_net) ** (1 / num_years1) - 1
        
        beginning_value_eps = sorted_groups['BasicEPS'].iloc[0]
        ending_value_eps = sorted_groups['BasicEPS'].iloc[-1]
        filtered_df2 = df[df['BasicEPS'].notnull()]
        num_years2 = len(filtered_df2)
        cagr_eps = (ending_value_eps / beginning_value_eps) ** (1 / num_years2) - 1
        
        income_quarterly_dataframe = income_quarterly_dataframe.append({
            'cagr_revenue_qr': cagr_rev,
            'ticker': i,
            'cagr_net_income_qr': cagr_net_in, # replace with appropriate calculation
            'cagr_eps_value_qr':cagr_eps # replace with appropriate calculation
        }, ignore_index=True)        
    return income_quarterly_dataframe

df_cagr = income_quarterly_Values_cagr(income)
df_cagr
```
